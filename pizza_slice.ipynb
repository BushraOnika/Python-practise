# What is a bigger pizza with a 25 inch radius or a square with 25 inch sides?

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc,Polygon
```

```python
radius=25
side=25
```

```python
round_pizza_area=3.1416*radius*radius
square_pizza_area=side*side
```

```python
round_pizza_area,square_pizza_area
```

```python
fig,axes= plt.subplots(figsize=(8,8))

circle=plt.Circle((0,0),25)
axes.set_aspect(1)
axes.add_patch(circle)
axes.set_xlim(-50,50)
axes.set_ylim(-50,50)
plt.show()
```

```python
fig, axes=plt.subplots()
square=plt.Rectangle((0,0),25,25)
axes.add_patch(square)
axes.set_xlim(-50,50)
axes.set_ylim(-50,50)
plt.show()
```

```python
fig , axes= plt.subplots()
circle=plt.Circle((0,0),25,color='pink')
square=plt.Rectangle((-12.5,-12.5),25,25,color='purple')
axes.add_patch(circle)
axes.add_patch(square)
axes.set_xlim(-30,30)
axes.set_ylim(-30,30)
axes.set_aspect('equal')
axes.set_title('Comparison of Pizza and Square (Overlay)')
plt.grid()
plt.show()
```

# Smiley Face


```python
fig, axes = plt.subplots()
circle=Circle((0,0),2,color='pink')
right_eye=Circle((-0.5,0.5),0.2)
left_eye=Circle((0.5,0.5),0.2)
smile=Arc((0,-0.5),1,1,theta1=200,theta2=340)
axes.add_patch(circle)
axes.add_patch(right_eye)
axes.add_patch(left_eye)
axes.add_patch(smile)
axes.set_xlim(-3,3)
axes.set_ylim(-3,3)
axes.set_aspect('equal')
```

# 4 faces together


```python
def set_limiltxy(axis,x,y):
    axis.set_xlim(-x,x)
    axis.set_ylim(-y,y)
    axis.set_aspect('equal')
```

```python
def make_smiley_face(axes):
    circle=Circle((0,0),2,color='pink')
    right_eye=Circle((-0.5,0.5),0.2)
    left_eye=Circle((0.5,0.5),0.2)
    smile=Arc((0,-0.5),1,1,theta1=200,theta2=340)
    axes.add_patch(circle)
    axes.add_patch(right_eye)
    axes.add_patch(left_eye)
    axes.add_patch(smile)
```

```python
fig , ((ax1,ax2),(ax3,ax4))=plt.subplots(2,2)
set_limiltxy(ax1,3,3)
make_smiley_face(ax1)
ax1.set_title('Abbu')
set_limiltxy(ax2,3,3)
make_smiley_face(ax2)
ax2.set_title('Ammu')
set_limiltxy(ax3,3,3)
make_smiley_face(ax3)
ax3.set_title('Onika')
set_limiltxy(ax4,3,3)
make_smiley_face(ax4)
ax4.set_title('Apurbo')
```

```python
fig,ax=plt.subplots()
map=Polygon([(1,1),(2,2),(3,1)])
ax.add_patch(map)
set_limiltxy(ax,4,4)
plt.show()
```
