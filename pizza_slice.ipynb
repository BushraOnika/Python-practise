# 4 faces together


```python
def set_limiltxy(axis,x,y):
    axis.set_xlim(-x,x)
    axis.set_ylim(-y,y)
    axis.set_aspect('equal')
```

```python
def make_smiley_face(axes):
    circle=Circle((0,0),2,color='pink')
    right_eye=Circle((-0.5,0.5),0.2)
    left_eye=Circle((0.5,0.5),0.2)
    smile=Arc((0,-0.5),1,1,theta1=200,theta2=340)
    axes.add_patch(circle)
    axes.add_patch(right_eye)
    axes.add_patch(left_eye)
    axes.add_patch(smile)
```

```python
fig , ((ax1,ax2),(ax3,ax4))=plt.subplots(2,2)
set_limiltxy(ax1,3,3)
make_smiley_face(ax1)
ax1.set_title('Abbu')
set_limiltxy(ax2,3,3)
make_smiley_face(ax2)
ax2.set_title('Ammu')
set_limiltxy(ax3,3,3)
make_smiley_face(ax3)
ax3.set_title('Onika')
set_limiltxy(ax4,3,3)
make_smiley_face(ax4)
ax4.set_title('Apurbo')
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc
import matplotlib.lines as mlines

def draw_detailed_face(ax, x, y, radius=0.4, face_color='peachpuff', label='', hair_style=None):
    # Head
    head = Circle((x, y), radius, facecolor=face_color, edgecolor='black', lw=2)
    ax.add_patch(head)

    eye_y = y + 0.15 * radius
    eye_dx = 0.15 * radius
    eye_radius = 0.05 * radius

    # Eyes
    for dx in [-eye_dx, eye_dx]:
        ax.add_patch(Circle((x + dx, eye_y), eye_radius, color='white', ec='black'))
        ax.add_patch(Circle((x + dx, eye_y), eye_radius/2, color='black'))

    # Eyebrows
    ax.plot([x - eye_dx - 0.05, x - eye_dx + 0.05],
            [eye_y + 0.07, eye_y + 0.07], color='black', lw=1)
    ax.plot([x + eye_dx - 0.05, x + eye_dx + 0.05],
            [eye_y + 0.07, eye_y + 0.07], color='black', lw=1)

    # Nose (simple line)
    ax.plot([x, x], [y + 0.05, y - 0.05], color='black', lw=1)

    # Mouth (smile arc)
    mouth = Arc((x, y - 0.1), 0.3 * radius * 2, 0.2 * radius * 2,
                theta1=0, theta2=180, color='black', lw=1.5)
    ax.add_patch(mouth)

    # Optional hair styles
    if hair_style == 'short':
        ax.plot([x - radius, x + radius], [y + radius, y + radius], color='brown', lw=3)
    elif hair_style == 'long':
        ax.add_patch(Arc((x, y + 0.05), radius*2.2, radius*2.5, theta1=0, theta2=180,
                         color='darkred', lw=4))
    elif hair_style == 'bun':
        ax.add_patch(Circle((x, y + radius + 0.1), 0.1, color='brown'))

    # Label
    if label:
        ax.text(x, y - radius - 0.2, label, ha='center', fontsize=10)

# Set up canvas
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_xlim(0, 10)
ax.set_ylim(0, 4)
ax.axis('off')

# Draw each family member
draw_detailed_face(ax, 2, 2.5, radius=0.5, label='Father', hair_style='short')
draw_detailed_face(ax, 4, 2.5, radius=0.45, label='Mother', hair_style='bun')
draw_detailed_face(ax, 6, 2.5, radius=0.35, label='Child 1', hair_style='short')
draw_detailed_face(ax, 8, 2.5, radius=0.33, label='Child 2', hair_style='long')

plt.title("Family of 4  Detailed Faces", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Circle, Arc
import matplotlib.lines as mlines

def draw_cute_face(ax, x, y, radius=0.4, face_color='#FFE0BD', label='', eye_color='black', blush_color='#FFC0CB'):
    # Face
    head = Circle((x, y), radius, facecolor=face_color, edgecolor='black', lw=1.5)
    ax.add_patch(head)

    # Eyes (big and sparkly)
    eye_y = y + 0.15 * radius
    eye_dx = 0.15 * radius
    eye_radius = 0.08 * radius
    sparkle_radius = eye_radius / 2.5

    for dx in [-eye_dx, eye_dx]:
        eye = Circle((x + dx, eye_y), eye_radius, color=eye_color)
        sparkle = Circle((x + dx - 0.02, eye_y + 0.02), sparkle_radius, color='white')
        ax.add_patch(eye)
        ax.add_patch(sparkle)

    # Smile (curved and friendly)
    smile = Arc((x, y - 0.05), 0.3 * radius * 2, 0.25 * radius * 2,
                theta1=0, theta2=180, color='black', lw=1.2)
    ax.add_patch(smile)

    # Cheeks (blush)
    blush_dx = 0.22 * radius
    blush_y = y - 0.02
    for dx in [-blush_dx, blush_dx]:
        cheek = Circle((x + dx, blush_y), 0.05 * radius, color=blush_color, alpha=0.6)
        ax.add_patch(cheek)

    # Label below
    ax.text(x, y - radius - 0.2, label, ha='center', fontsize=10)

# Set up canvas
fig, ax = plt.subplots(figsize=(10, 5))
ax.set_xlim(0, 10)
ax.set_ylim(0, 4)
ax.axis('off')

# Draw cute family
draw_cute_face(ax, 2, 2.5, radius=0.5, label='Father')
draw_cute_face(ax, 4, 2.5, radius=0.45, label='Mother')
draw_cute_face(ax, 6, 2.5, radius=0.35, label='Child 1')
draw_cute_face(ax, 8, 2.5, radius=0.33, label='Child 2')

plt.title("Cute & Smiley Cartoon Family Faces", fontsize=16)
plt.show()
```

```python
import time

def timer(func):
    def wrapper(*args, **kwargs):
        start_time = time.perf_counter_ns()
        result = func(*args, **kwargs)
        end_time = time.perf_counter_ns()
        eclipsed_time = end_time - start_time
        print(f"method took time: {eclipsed_time:.4f}ns")
        return result

    return wrapper

@timer
def add(a, b):
    return a + b

add(1, 2)
```
